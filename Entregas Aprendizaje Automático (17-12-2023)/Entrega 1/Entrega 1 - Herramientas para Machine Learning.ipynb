# Construcción del modelo y elección para tener la exactitud del 85%
## Split
Se separan las variables independientes y dependientes del DataFrame para empezar a trabajar en el entrenamiento con los diferentes modelos.

```python
X = np.array(df.drop(['VF'],1))
y = np.array(df['VF'])
X.shape
```

Se realiza el escalado/normalizado del DataFrame

```python
mms = StandardScaler()
X_mms = mms.fit_transform(X)
```

Se realiza la subdivisión del dataset en un set de entrenamiento y uno de validación de los modelos, para poder preparar el entrenamiento. Se definen los **parámetros del tamaño** de cada set (20% para validar, el resto para entrenar).
- Se selecciona una semilla aleatoria para reproducir la selección. Se selecciona el número 42 al ser de uso común en Machine Learning.

```python
validation_size = 0.2
seed = 42
train_X, val_X, train_y, val_y = model_selection.train_test_split(X_mms, y, test_size=validation_size, random_state=seed)
```

# Modelo 1. Logistic Regression
Se inicia el entrenamiento del primer modelo, **Logistic Regression**

```python
model = linear_model.LogisticRegression()
model.fit(train_X,train_y)
```

Se prueba el modelo de cross validation con ls valores por defecto del mismo, utilizando nuestras muestras previamente definidas.
- Se obtiene la precisión del modelo al realizar **5 folds de los datos de entrenamiento** (80% del dataset) obteniendo diferentes valores de precisión en un array por cada fold, y al final sacando la media de los mismos, obteniendo una precisión alta del 97% para nuestro modelo.

```python
kfold = model_selection.KFold(n_splits=5)

cv_results = model_selection.cross_val_score(model, train_X, train_y,
                                             cv=kfold,
                                             scoring='accuracy')

print(cv_results.mean())
```

```python
pred = model.predict(val_X)
print(accuracy_score(val_y, pred))
```

El modelo asigna una probabilidad a cada clase ("0" ó "1") para cada predicción, de acuerdo con su entrenamiento.
- La variable con la probabilidad más alta es la que el modelo predice. Cuanto más alta sea la probabilidad de una clase, más probable es que la predicción del modelo que generamos sea correcta.

```python
prob = model.predict_proba(val_X)
np.set_printoptions(formatter={'float_kind': '{:.4f}'.format})
print(prob)
```

Se revisa la clase proporcionada en base a la probabilidad.

```python
pred = model.predict(val_X)
np.set_printoptions(threshold=np.inf)
print(pred)
```

Se genera la **matriz de confusión** respectiva para determinar la cantidad de resultados equivocados que se generaron en nuestro modelo.

```python
print(confusion_matrix(val_y, pred))
```

### ¿Que se puede interpretar en la matriz de confusión?
- Al evaluar las dos variables de la matriz, donde **cero** es la inexistencia de la anomalía cardiaca, y **uno** es la existencia de la misma.
 - Podemos ver que la clase 1 (cero, sin anomalía) se predijo correctamente **1518 veces** y se predijo erróneamente (como *clase 2*) **18 veces**
 - Por otro lado, para la clase 2 (uno, anomalía cardiaca) se predijo correctamente **275 veces**, y se predijo erróneamente (como *clase 1*) ***19 veces***.
 - Al poder observar la proporción de errores contra aciertos del modelo en la matriz de confusión, se puede deducir que este modelo predice con mayor certeza la clase 1 (inexistencia de anomalía) que la clase 2 (existencia de anomalía), datos que podrían mejorarse al utilizar otros modelos.

Se genera el F1-score para conocer el ajuste del modelo generado.

```python
print(classification_report(val_y, pred))
```

La exactitud promedio del modelo en F1 **se encuentra en 98%**, por lo que se puede considerar que nuestra validación presenta un **excelente porcentaje de ajuste**, aunque si se evalúa particularmente a las anomalías cardiacas (variable "1" o clase 2) este porcentaje de F1 cae al 94%.

# Modelo 2. KNN
Para comprobar lo mencionado anteriormente y ver si se puede incrementar la exactitud, se probará ahora el modelo **K Nearest Neighbors (vecinos más cercanos - KNN)**, siguiendo los mismos pasos anteriores pero seleccionando este modelo.

```python
model_knn = KNeighborsClassifier()
model_knn.fit(train_X,train_y)

kfold = model_selection.KFold(n_splits=5)

cv_results = model_selection.cross_val_score(model_knn, train_X, train_y,
                                             cv=kfold,
                                             scoring='accuracy')

print(cv_results.mean())

pred_knn = model_knn.predict(val_X)
print(accuracy_score(val_y, pred_knn))

prob = model_knn.predict_proba(val_X)
#print(prob)

print(confusion_matrix(val_y, pred_knn))

print(classification_report(val_y, pred_knn))
```

Como se puede apreciar en la matriz de confusión, el modelo de vecino más cercano (KNN) sí mejora la precisión para los casos de inexistencia de anomalía cardiaca, pero **incrementa ligeramente los falsos negativos en caso de la existencia de la misma**.

# Modelo 3. Random Forest
Ahora se probará el modelo **Random Forest (bosque aleatorio)**, siguiendo los pasos anteriores, pero cambiando este modelo, para ver si mejora la exactitud.

```python
model_rf = RandomForestClassifier()
model_rf.fit(train_X,train_y)

kfold = model_selection.KFold(n_splits=5)

cv_results = model_selection.cross_val_score(model_rf, train_X, train_y,
                                             cv=kfold,
                                             scoring='accuracy')

print(cv_results.mean())

pred_rf = model_rf.predict(val_X)
print(accuracy_score(val_y, pred_rf))

prob = model_rf.predict_proba(val_X)
#print(prob)

print(confusion_matrix(val_y, pred_rf))

print(classification_report(val_y, pred_rf))
```

A diferencia de lo que sucedió en el modelo anterior, y al observar la matriz de confusión, para el modelo de Random Forest si vemos **una mejora en la exactitud en ambas variables**, alcanzando incluso una exactitud promedio del 99% en el modelo, y de 96% para las anomalías cardiacas ("1").

# Modelo 4. XGBoost
Ahora se probará el modelo **XGBoost**, siguiendo los pasos anteriores, pero cambiando este modelo, para ver las variaciones en la exactitud de ambas clases.

```python
model_xg = XGBClassifier()
model_xg.fit(train_X,train_y)

kfold = model_selection.KFold(n_splits=5)

cv_results = model_selection.cross_val_score(model_xg, train_X, train_y,
                                             cv=kfold,
                                             scoring='accuracy')

print(cv_results.mean())

pred_xg = model_xg.predict(val_X)
print(accuracy_score(val_y, pred_xg))

prob = model_xg.predict_proba(val_X)
#print(prob)

print(confusion_matrix(val_y, pred_xg))

print(classification_report(val_y, pred_xg))
```

Para el caso del modelo XGBoost, podemos ver que si logra incrementar la exactitud para las anomalías cardiacas ("1"), sin embargo, su exactitud **disminuye ligeramente** para la ausencia de las mismas ("0"). El análisis posterior de cada modelo permitirá determinar cual de los dos modelos podría ser el más idóneo para predecir las situaciones de nuestro interés en un conjunto real de datos.
