# Contexto

Vamos a utilizar un conjunto de datos sobre vinos con sus características para crear un modelo de aprendizaje automático extraidos de la web https://www.vivino.com/ES/es/.

Vivino.com permite a los usuarios buscar y explorar una amplia variedad de vinos, leer reseñas y calificaciones de otros usuarios, y mantener un registro de los vinos que han probado y les han gustado.

Algunas de las características comunes de Vivino.com incluyen:

- Búsqueda y exploración de vinos: Los usuarios pueden buscar vinos por nombre, tipo de uva, región vinícola y otros criterios.
- Calificaciones y reseñas: Los usuarios pueden calificar y escribir reseñas de vinos que han probado, lo que proporciona información valiosa para otros amantes del vino.
- Escaneo de etiquetas: Una característica distintiva de Vivino es su capacidad para escanear etiquetas de vino con la cámara del teléfono y proporcionar información instantánea sobre ese vino, incluyendo calificaciones y reseñas de otros usuarios.
- Recomendaciones personalizadas: Vivino utiliza algoritmos para ofrecer recomendaciones de vinos basadas en las preferencias de los usuarios y las calificaciones que han dado a otros vinos.
- Listas de vinos y seguimiento de bodega: Los usuarios pueden crear listas de vinos que deseen probar o seguir a sus bodegas favoritas.
- Compra de vinos: Vivino permite a los usuarios comprar vinos directamente a través de la plataforma en algunos mercados.

# Contenido del fichero

Los datos contienen tipos de vino: tinto, blanco, rosado y espumoso y son extraidos utilizando web-scraping para acceder al contenido de las siguientes variables:

- Name: Nombre del vino

### Atípicos

#### Por desviación típica

##### Identificación

###### Definir el número de desviaciones típicas

```python
num_desv_tip = 4
```

###### Función que devuelve los índices de los atípicos

```python
def atipicos_desv_tip(variable, num_desv_tip = 4):
    #sacamos los nulos por ahora
    variable = variable.dropna()
    #calculamos los límites
    media = np.mean(variable)
    sd = np.std(variable)
    umbral = sd * num_desv_tip
    lim_inf = media - umbral
    lim_sup = media + umbral
    #encontramos los índices de los que están fuera de los límites
    indices = [indice for indice,valor in variable.items() if valor < lim_inf or valor > lim_sup]
    return(indices)
```

###### Función que cuenta el número de atípicos

```python
df.shape[0]
```

```python
def conteo_atipicos(df,variable, num_desv_tip = 4):
    atipicos = atipicos_desv_tip(df[variable], num_desv_tip)
    print('% de Observaciones fuera de límites: ',len(atipicos)/df.shape[0]*100)
    return(df.loc[atipicos,variable].value_counts())
```

###### Variables a aplicar

```python
var_atipicos_dt = ['rating','numberofratings','price']
```

###### Aplicar

```python
for variable in var_atipicos_dt:
    print('\n' + variable + ':\n')
    print(conteo_atipicos(num,variable,num_desv_tip))
```

### Análisis estadístico

```python
def frecuencias_cat(df_cat):
    resultado = df_cat.apply(lambda x: x.value_counts()).T.stack()\
                .to_frame().reset_index()\
                .rename(columns={'level_0':'Variable','level_1':'Valor',0: "Frecuencia"})\
                .sort_values(by = ['Variable','Frecuencia'])
    return(resultado)
```

### Análisis gráfico

```python
def graficos_eda_categoricos(cat):
    
    #Calculamos el número de filas que necesitamos
    from math import ceil
    filas = 2

    #Definimos el gráfico
    f, ax = plt.subplots(nrows = filas, ncols = 2, figsize = (16, filas * 6))

    #Aplanamos para iterar por el gráfico como si fuera de 1 dimensión en lugar de 2
    ax = ax.flat 

    #Creamos el bucle que va añadiendo gráficos
    for cada, variable in enumerate(cat):
        cat[variable].value_counts().plot.barh(ax = ax[cada])
        ax[cada].set_title(variable, fontsize = 12, fontweight = "bold")
        ax[cada].tick_params(labelsize = 12)
```

## EDA NUMÉRICAS

### Análisis estadístico

```python
def estadisticos_cont(num):
    #Calculamos describe
    estadisticos = num.describe().T
    #Añadimos la mediana
    estadisticos['median'] = num.median()
    #Reordenamos para que la mediana esté al lado de la media
    estadisticos = estadisticos.iloc[:,[0,1,8,2,3,4,5,6,7]]
    #Lo devolvemos
    return(estadisticos)
```

```python
estadisticos_cont(num)
```

### Análisis gráfico

```python
def graficos_eda_continuas(num):
    
    #Calculamos el número de fila que necesitamos
    from math import ceil
    filas = ceil(num.shape[1] / 2)

    #Definimos el gráfico
    f, ax = plt.subplots(nrows = filas, ncols = 2, figsize = (16, filas * 6))

    #Aplanamos para iterar por el gráfico como si fuera de 1 dimensión en lugar de 2
    ax = ax.flat 

    #Creamos el bucle que va añadiendo gráficos
    for cada, variable in enumerate(num):
        num[variable].plot.density(ax = ax[cada])
        ax[cada].set_title(variable, fontsize = 12, fontweight = "bold")
        ax[cada].tick_params(labelsize = 12)
```

## IMPORTAR PAQUETES

```python

from sklearn.model_selection import train_test_split

from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.metrics import calinski_harabasz_score
from sklearn.metrics import davies_bouldin_score

from sklearn.model_selection import GridSearchCV

#Autocompletar rápido
%config IPCompleter.greedy=True

#Desactivar la notación científica
pd.options.display.float_format = '{:.2f}'.format
```

```python
df.columns
```

```python
df_c = df[['country_Argentina', 'country_Australia',
       'country_Austria', 'country_Brazil', 'country_Bulgaria',
       'country_Canada', 'country_Chile', 'country_China', 'country_Croatia',
       'country_Czech Republic', 'country_France', 'country_Georgia',
       'country_Germany', 'country_Greece', 'country_Hungary',
       'country_Israel', 'country_Italy', 'country_Lebanon',
       'country_Luxembourg', 'country_Mexico', 'country_Moldova',
       'country_New Zealand', 'country_Portugal', 'country_Romania',
       'country_Slovakia', 'country_Slovenia', 'country_South Africa',
       'country_Spain', 'country_Switzerland', 'country_Turkey',
       'country_United Kingdom', 'country_United States', 'country_Uruguay',
       'wine_type_blanco', 'wine_type_espumoso', 'wine_type_rosado',
       'wine_type_tinto', 'rating_mms', 'numberofratings_mms', 'price_mms','year']]
```

### Crear el espacio de soluciones

```python
min_k = 2
max_k = 10

soluciones = list(range(min_k,max_k))
```

### Crear listas para guardar la salida de cada métrica

```python
codo = []
silueta = []
calins = []
davies = []
```

### Analizar el espacio de soluciones

```python
corrs = df_c[['rating_mms','numberofratings_mms','price_mms']].corr() #Heatmap for numetrical columns
fig, ax = plt.subplots(figsize=(7,5))        

sns.heatmap(corrs,annot = True,ax=ax,linewidths=.6, cmap = 'YlGnBu');
```

```python
for solucion in soluciones:
    
    #Instanciar
    cluster = KMeans(n_clusters = solucion, n_init=10)
    
    #Entrenar
    cluster.fit(df_c)
    
    #Recoger las métricas
    codo.append(cluster.inertia_)
    silueta.append(silhouette_score(df_c, cluster.labels_))
    calins.append(calinski_harabasz_score(df_c, cluster.labels_))
    davies.append(davies_bouldin_score(df_c, cluster.labels_))
```

### PCA

```python
pca = PCA(n_components=df_c.shape[1])
pca_model = pca.fit(df_c)
var = pca_model.explained_variance_ratio_
print(np.around(var, 3))
print('')
principalComponents = pca_model.transform(df_c)
#print(principalComponents[:,1])
cumulative_var = np.cumsum(np.around(var, 3)*100)
print(cumulative_var)

plt.figure(figsize=(10,8))
plt.plot(cumulative_var)
#plt.ylim(0,30)
plt.title("Principal Component Analysis")
plt.xlabel("No. of Components")
plt.ylabel("Cumulative Proportion of Variance Explained")
plt.show()
```

```python
plt.figure(figsize=(12,8))

x=principalComponents[:, 0]
y=principalComponents[:, 1]

plt.figure(figsize=(12,8))
sns.scatterplot(data=df,x=x, y=y, hue=df.wine_type,palette="Set1",size="price",sizes=(20, 200))
plt.title('PCA 2D Posicion')
plt.xlabel('CP1')
plt.ylabel('CP2')
plt.show();
```

```python
plt.figure(figsize=(12,8))

x=principalComponents[:, 0]
y=principalComponents[:, 1]

plt.figure(figsize=(12,8))
sns.scatterplot(data=df,x=x, y=y, hue=df.country,palette="Set1",size="price",sizes=(20, 200))
plt.title('PCA 2D Posicion')
plt.xlabel('CP1')
plt.ylabel('CP2')
plt.show();
```

### Identificar la mejor solucion

#### Analizar las métricas

```python
metricas = pd.DataFrame({'Solucion':soluciones,
              'Codo': codo,
              'Silueta':silueta,
              'Calins':calins,           
              'Davies':davies}).set_index('Solucion')

metricas
```

#### Analizar los gráficos

```python
metricas.plot(subplots = True, figsize = (12,8), layout = (2,2), sharex = False);
```

### Segmentación final con la mejor solucion

```python
mejor_solucion = 5

#Instanciar
cluster = KMeans(n_clusters = mejor_solucion)
    
#Entrenar
cluster.fit(df_c)
```
